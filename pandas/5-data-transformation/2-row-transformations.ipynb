## row transformation in Pandas
- adding rows
- deleting rows
- swapping rows


```python
# Create sample DataFrame
df = pd.DataFrame({
        'A': [1, 2, 3, 4],
        'B': [5, 6, 7, 8],
        'C': [9, 10, 11, 12]
})
df
```

```python
# Pandas does not have a built-in insert() operation for rows 
# like it does for columns (df.insert()).
# Note that row insertion breaks the RangeIndex, so not efficient

new_row =  {'A': 10, 'B': 11, 'C': 12}
position=2   # One before the end

print('index: ', df.index)
df1 = pd.concat([
    df.iloc[:position], 
    pd.DataFrame([new_row]), 
    df.iloc[position:]
])
print(df1)
print('index: ', df1.index)
df1 = df1.reset_index(drop=True)  # âœ— No built-in method
print('index: ', df1.index)

df1
```

```python
print('index: ', df.index)
df3 = df.drop(index=1)  # Deletes row with index 1
print('index: ', df3.index)
# Let's test the df in this phase:
print('loc: \n', df3.loc[2])      # This uses the location from the Index
print('iloc: \n', df3.iloc[2])    # This uses the true location

df3 = df3.reset_index(drop=True)  # Reset index
print('index: ', df3.index)
   
df3
```

```python
# Swap first and last rows

df5 = df.copy()
df5.iloc[0], df5.iloc[-1] = df5.iloc[-1].copy(), df5.iloc[0].copy()

df5
```

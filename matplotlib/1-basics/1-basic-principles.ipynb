```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Standard pattern used in data science:
# 1. Create figure and axes with plt.subplots()
fig, ax = plt.subplots(figsize=(10, 6))
print('figure: ', type(fig))
print('axes: ', type(ax))
```

```python

# 2. Use axes methods for everything else

ax.plot([1, 2, 3], [1, 2, 3])
ax.set_title('Basic Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

# Don't use these pyplot commands when possible:
# plt.plot()
# plt.title()
# plt.xlabel()
```

```python
# Create sample data
dates = pd.date_range('2023-01-01', periods=100)
values = np.random.randn(100).cumsum()

# Standard plotting pattern
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(dates, values)
ax.set_title('Time Series')

# Adding multiple plots to same axes
ax.plot(dates, values + 2, label='Series 2')
ax.plot(dates, values - 2, label='Series 3')
ax.legend()
ax.grid(True)
```

This will add the ability of matplotlib to display the interactive plot.

````python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

print('backend: ', plt.get_backend())
print('-----------------------------')
# Create plot
fig, ax = plt.subplots(figsize=(8, 6))
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.grid(True)
ax.set_title('Interactive Sine Wave')
````

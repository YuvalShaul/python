This will add the ability of matplotlib to display the interactive plot.

````python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

print('backend: ', plt.get_backend())
print('-----------------------------')
# Create plot
fig, ax = plt.subplots(figsize=(8, 6))
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.grid(True)
ax.set_title('Interactive Sine Wave')
````

At the bottom of your plot, you'll see a toolbar with these buttons:
 Home - Resets the view to the original zoom/pan state
 Back /  Forward - Navigate through your zoom/pan history (like browser back/forward)
 Pan - Click this, then click and drag to move around the plot

Useful when zoomed in to explore different areas

 Zoom - Click this, then:

- Left-click and drag to zoom into a rectangular area
- Right-click to zoom out

 Configure - Opens subplot configuration dialog

- Adjust margins, spacing between subplots
- Mostly for multi-subplot figures

 Save - Download the current plot as PNG

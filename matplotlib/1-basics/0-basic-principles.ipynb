```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Working with matplotlib - approaches
There are 2 approaches generally used to work with matplotlib, as described [here](https://matplotlib.org/stable/users/explain/figure/api_interfaces.html):
- An explicit "Axes" interface that uses methods on a Figure or Axes object to create other Artists, and build a visualization step by step. This has also been called an "object-oriented" interface.
- An implicit "pyplot" interface that keeps track of the last Figure and Axes created, and adds Artists to the object it thinks the user wants.

In this course we are generally taking the 2st approach, but can call Artist methods directly if we need to. So after creating a figure and axes objects, we will use axes methods directly to do things.

```python
# Implicit approach:
# We do not even create a figure and axes objects.
# The pyplot module create it implicitly, and keep track of these objects.
# They become the "current" figure and axes
print(plt.get_backend())
# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot using implicit approach
plt.plot(x, y)

# Add labels and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Line Chart')

# Add grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
# More explicit approach (still uses PyPlot):
# We create a figure and axes objects explicitly
# We call all plotting methods directly from the correct axes object

# Let's create some random data:
dates = pd.date_range('2023-01-01', periods=100)
values = np.random.randn(100).cumsum()

# 1. Create figure and axes with plt.subplots()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(dates, values, label='first plot')
ax.set_title('Time Series')


# 2. Use axes methods for everything else
# Adding multiple plots to same axes
ax.plot(dates, values + 2, label='Series 2')
ax.plot(dates, values - 2, label='Series 3')
ax.legend()
ax.grid(True)

# Don't use these pyplot commands when possible:
# plt.plot()
# plt.title()
# plt.xlabel()
# Bacause:  
#    pyplot maintains a "current figure" and "current axes" behind the scenes
#    It is Harder to control: When you have multiple plots, it's unclear which one you're modifying
```

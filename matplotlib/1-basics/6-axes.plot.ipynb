```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a figure and axis object
fig, ax = plt.subplots()

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic plotting using axes
ax.plot(x, y, label='Sine wave')

# Customize the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Simple Axes Plot')
ax.grid(True)
ax.legend()

plt.show()
```

### How to use the *arg parameter in matplotlib.pyplot.plot()

```python
# Create example data
x = np.linspace(0, 10, 10)
y1 = x**2
y2 = x**3

fig, ax = plt.subplots()

# 1. Y values only (x is auto-generated as indices)
ax.plot(y1)

# 2. X and Y values
ax.plot(x, y1)

# 3. Multiple plots in one call
ax.plot(x, y1, x, y2)

# 4. Including format strings
ax.plot(x, y1, 'r--')  # red dashed line
ax.plot(x, y1, 'go')   # green dots
ax.plot(x, y1, 'b-', x, y2, 'r:')  # blue solid line and red dotted line

# 5. Single format string for multiple plots
ax.plot(x, y1, x, y2, 'r--^')  # both lines will be red dashed, with triangles
```

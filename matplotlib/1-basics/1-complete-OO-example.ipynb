Note that I do not import pyplot

```python
from matplotlib.figure import Figure
from matplotlib import rcParams         # These are the Runtime Configuration Params
from IPython import display
import numpy as np
from datetime import datetime, timedelta

# I want to know what backend I am using
print(rcParams['backend'])

# Create data
dates = [datetime(2024, 1, 1) + timedelta(days=x) for x in range(100)]
values = np.random.randn(100).cumsum()
values2 = np.random.randn(100).cumsum()

# Create Figure instance directly
fig = Figure(figsize=(12, 8))

# Create GridSpec for layout
gs = fig.add_gridspec(2, 2)

# Main time series plot
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(dates, values, label='Series 1')
ax1.plot(dates, values2, label='Series 2')
ax1.set_title('Time Series Analysis')
ax1.set_xlabel('Date')
ax1.set_ylabel('Value')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Supporting plots
ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(values, bins=30, alpha=0.6, label='Series 1')
ax2.hist(values2, bins=30, alpha=0.6, label='Series 2')
ax2.set_title('Distribution')
ax2.legend()

ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(values, values2, alpha=0.5)
ax3.set_title('Correlation')
ax3.set_xlabel('Series 1')
ax3.set_ylabel('Series 2')

# Adjust layout
fig.tight_layout()

# Display
display.display(fig)
```

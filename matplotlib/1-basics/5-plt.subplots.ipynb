### Some plt.subplots examples

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Example 1: Basic 2x2 subplot grid
# Parameters: 2x2 array of subfigures (axes objects) and figure size in inches
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
x = np.linspace(0, 10, 100)

# Plot different functions in each subplot
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sine Wave')

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('Cosine Wave')

axs[1, 0].plot(x, x**2)
axs[1, 0].set_title('Quadratic Function')

axs[1, 1].plot(x, np.exp(x/3))
axs[1, 1].set_title('Exponential Function')

plt.tight_layout()
plt.show()
```

```python
# Example 2: Subplots with shared axes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), 
                              sharex=True,  # Share x-axis between subplots
                              height_ratios=[2, 1])  # First plot twice as tall

x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x))
ax1.set_title('Main Plot')

ax2.plot(x, np.cos(x))
ax2.set_title('Secondary Plot')

plt.tight_layout()
plt.show()
```

```python
# Example 3: Subplots with spacing and grid options
# customize spacing between subplots and create subplots of different sizes using gridspec_kw

fig, axs = plt.subplots(2, 2, figsize=(10, 8),
                        gridspec_kw={
                            'hspace': 0.4,  # Vertical space between plots
                            'wspace': 0.3,  # Horizontal space between plots
                            'width_ratios': [1.5, 1],  # First column 1.5x wider
                            'height_ratios': [1, 1.2]  # Second row 1.2x taller
                        })

# Add some example plots
for i in range(2):
    for j in range(2):
        axs[i, j].plot(np.random.rand(10))
        axs[i, j].set_title(f'Plot {i+1},{j+1}')
        axs[i, j].grid(True)  # Add grid to all subplots

plt.show()
```

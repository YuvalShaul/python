```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create figure and axes - 2 rows, 2 columns
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
plt.close()  # Prevents automatic display
```

```python
# Create some sample data
x = np.linspace(0, 10, 100)

# Now axes is a 2D array of Axes objects
# Can access them as axes[row, col]

# Top left - sine wave
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Sine')

# Top right - cosine wave
axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title('Cosine')

# Bottom left - parabola
axes[1, 0].plot(x, x**2)
axes[1, 0].set_title('Parabola')

# Bottom right - exponential
axes[1, 1].plot(x, np.exp(x/3))
axes[1, 1].set_title('Exponential')
```

```python
# Add some spacing between subplots
# Cell 3
from IPython.display import display
fig.tight_layout()
display(fig)
```

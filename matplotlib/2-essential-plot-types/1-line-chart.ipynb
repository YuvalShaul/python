```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Chart
From [wikipedia - line chart](https://en.wikipedia.org/wiki/Line_chart):
"A line chart or line graph, also known as curve chart, is a type of chart that displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields."

```python
# Create sample data - two different series
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axis objects with specific figure size
fig, ax = plt.subplots(figsize=(10, 6))

# Plot both lines with labels and different styles
ax.plot(x, y1, label='Sine', color='#1f77b4', linestyle='-', linewidth=2)
ax.plot(x, y2, label='Cosine', color='#ff7f0e', linestyle='--', linewidth=2)

# Customize the plot
ax.set_xlabel('Time', fontsize=12)
ax.set_ylabel('Amplitude', fontsize=12)
ax.set_title('Sine and Cosine Waves', fontsize=14, pad=15)
ax.grid(True, linestyle=':', alpha=0.7)

# Customize axis properties
ax.set_xlim(0, 10)    # set limits for x and y axes
ax.set_ylim(-1.5, 1.5)

# Add legend
ax.legend(loc='upper right', frameon=True, shadow=True)

# Customize tick labels
ax.tick_params(axis='both', which='major', labelsize=10)

# Add a text box with some information
ax.text(0.02, 0.98, 'Period = 2', 
        transform=ax.transAxes,
        bbox=dict(facecolor='white', alpha=0.8),
        verticalalignment='top')

plt.tight_layout()
plt.show()
```

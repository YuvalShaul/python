```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data
categories = ['Python', 'Java', 'JavaScript', 'C++', 'Ruby']
usage = [68, 45, 61, 35, 28]

# Create figure and axis
fig,ax = plt.subplots(figsize=(10, 6))

# Create basic bar chart
ax.bar(categories,      # X-axis categories
        usage,          # Heights of bars
        color='skyblue',
        edgecolor='black',
        width=0.6)      # Width of bars

# Customize the plot
fig.suptitle('Programming Language Usage', fontsize=14)
ax.set_xlabel('Language', fontsize=12)
ax.set_ylabel('Usage (%)', fontsize=12)

# Add value labels on top of each bar
for i, v in enumerate(usage):
    plt.text(i, v + 1, str(v), 
             ha='center',     # Horizontal alignment
             va='bottom')     # Vertical alignment

# Adjust layout
fig.tight_layout()
# Don't use fig.show(), as it is trying to use a non-interractive backend
# Jupyer will discover the figure and display it
```

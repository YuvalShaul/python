### Explicit (OO) pyplot simple example
- plt.figure() ([see documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html#matplotlib.pyplot.figure)) creates a new figure
- The fig.subplots() ([see documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.subplots.html)) adds a single axes subplot (actually a grid (1x1) of subplots) to the figure
- The axes.plot() ([see documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html))  Plot y versus x as lines and/or markers.

```python
import matplotlib.pyplot as plt

fig = plt.figure()
print(type(fig))
ax = fig.subplots()
print(type(ax))
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Basic linspace
x = np.linspace(0, 10, 5)  # 5 evenly spaced numbers from 0 to 10
print("Basic linspace:", x)  # [0. 2.5 5. 7.5 10.]
```

```python
# With endpoint=False
x_no_end = np.linspace(0, 10, 5, endpoint=False)
print("\nWithout endpoint:", x_no_end)  # [0. 2. 4. 6. 8.]
```

```python
# Common use case: plotting a sine wave
# linspace is used to create the x values
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python
# Get step size
x, step = np.linspace(0, 10, 5, retstep=True)
print(f"\nStep size: {step}")
```

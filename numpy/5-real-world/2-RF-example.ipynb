```python
import numpy as np
import matplotlib.pyplot as plt
```

### Signal Generation: 
Creates a 1 kHz sine wave sampled at 10 kHz for 10 milliseconds using NumPy's vectorized operations.

```python
# Define signal parameters
frequency = 1000  # 1 kHz
sample_rate = 10000  # 10 kHz
duration = 0.01  # 10 milliseconds

# Create time array
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Generate sine wave
signal = np.sin(2 * np.pi * frequency * t)
signal
```

### Power Calculation: 
Computes the average power of the signal (mean square value), which is an important RF parameter.

```python
# Compute signal power
power = np.mean(signal**2)
print(f"Signal power: {power:.4f}")
```

### Frequency Analysis: 
Performs a Fast Fourier Transform (FFT) to convert the signal from time domain to frequency domain, allowing visualization of the frequency components.

```python
# Perform Fast Fourier Transform (FFT) to view in frequency domain
signal_fft = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), 1/sample_rate)
```

### Plot time domain

```python
# Plot time domain signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t * 1000, signal)  # Convert to milliseconds for display
plt.title('Time Domain Signal')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.grid(True)
```

### Plot frequency domain

```python
# Plot frequency domain (only positive frequencies)
plt.subplot(2, 1, 2)
plt.plot(frequencies[:len(frequencies)//2], np.abs(signal_fft)[:len(frequencies)//2])
plt.title('Frequency Domain')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
